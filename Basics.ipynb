# Finding the Best Two Markets to Advertise In

Scenario: we're working for an an e-learning company that offers courses on programming. Most of our courses are on web and mobile development, but we also cover many other domains, like data science, game development, etc. We want to promote our product and we'd like to invest some money in advertisement. Our goal in this project is to find out the two best markets to advertise our product in.

## Understanding the Data

To reach our goal, we could organize surveys for a couple of different markets to find out which would the best choices for advertising. This is very costly, however, and it's a good call to explore cheaper options first.

We can try to search existing data that might be relevant for our purpose. One good candidate is the data from [freeCodeCamp's 2017 New Coder Survey](https://medium.freecodecamp.org/we-asked-20-000-people-who-they-are-and-how-theyre-learning-to-code-fff5d668969). [freeCodeCamp](https://www.freecodecamp.org/) is a free e-learning platform that offers courses on web development. Because they run [a popular Medium publication](https://medium.freecodecamp.org/) (over 400,000 followers), their survey attracted new coders with varying interests (not only web development), which is ideal for the purpose of our analysis.

The survey data is publicly available in [this GitHub repository](https://github.com/freeCodeCamp/2017-new-coder-survey).

```python
import pandas as pd

survey_data = pd.read_csv("2017-fCC-New-Coders-Survey-Data.csv")
print(survey_data.shape) # Print number of rows and columns
pd.options.display.max_columns = 150 # avoiding shortened visual output
survey_data.head()
```

## Checking for Sample Representativity

As we mentioned earlier, most of the courses we offer are on web and mobile development, but we also cover many other domains, like data science, game development, etc. For the purpose of our analysis, we want to answer questions about a population of new coders that are interested in the subjects we teach. We'd like to know:

- Where are these new coders located.
- What are the locations with the greatest number of new coders.
- How much money new coders are willing to spend on learning.

So, before starting to analyze the sample data we have, we need to clarify whether it's representative for our population of interest and it has the right categories of people for our purpose.

The `JobRoleInterest` column describes, for every participant, the role(s) they're interested in working in; so we can infer that these are the domains/topics that they'd be interested in learning about.  We can take a look at the frequency distribution of role interest and see if the information available is relevant to us:

```python
survey_data["JobRoleInterest"].value_counts(normalize = True) * 100 # Normalize and multiply by 100 for percentage value
```

Taking an initial look at our frequency table here, a couple of things are noticeable:

- Quite a few people are interested in web development of some form (either full-stack web development, back-end web development, or front-end web development).
- A few people are interested in mobile development.
- Quite a few people appear to have marked interest in more than one role.

To get a better idea bout the number of respondents who have marked a single role of interest versus multiple options, we can split the responses and take a look at the number of options selected by each.  We'll first drop null values from the column, since we can't split those.

```python
role_interests_clean = survey_data["JobRoleInterest"].dropna()
interests_split = role_interests_clean.str.split(",")
```

Now we can take a look at a frequency table that describes how many options respondents have selected:

```python
options_selected = interests_split.apply(lambda x: len(x))
options_selected.value_counts(normalize = True).sort_index() * 100
```

It turns out that only about 31.7% of respondents indicated a clear interest in a single role, while the majority indicated mixed interests.  Because we offer courses on multiple topics, mixed interests can still be valuable for us.

Given that the focus of our courses is web and mobile development, let's take a closer look at the number of respondents that chose at least one of those two options:

```python
web_mobile = role_interests_clean.str.contains('Web Developer|Mobile Developer')
web_mobile_freq = web_mobile.value_counts(normalize = True) * 100
print(web_mobile_freq)
```

```python
# Graph to visualize the above frequency table

import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use("fivethirtyeight")

web_mobile_freq.plot.bar(figsize=(10, 6))
plt.ylabel("Percentage")
plt.xticks([0, 1], ['Web or Mobile Development', 'Other'], rotation = 0)
plt.title("Percentage of Respondents Interested in\nWeb/Mobile Development", y = 1.1)
plt.ylim([0, 100])
plt.show()
```

We can see that a vast majority (roughly 86%) of participants indicated an interest in web or mobile development.  As a result, we have good reason to consider this dataset to be a representative sample of our population of interest.

Now, we can move forward in answering our target questions:

- Where are these new coders located.
- What are the locations with the greatest number of new coders.
- How much money new coders are willing to spend on learning.

## New Coders - Locations and Densities

Now that we found out that the sample has the right categories of people for our purpose, we can begin analyzing it. We can start with finding out where these new coders are located, and what the densities are (how many coderes there are) for each location.

The data set provides information about the location of each participant at a country level. The `CountryCitizen` variable describes the country of origin for each participant, and the `CountryLive` variable describes what country each participant lives in (which may be different from the origin country).

For our analysis, we'll work with the `CountryLive` variable because we're interested in where people are actually living at the moment we run the ads. In other words, we're interested where people are currently located, not where they were born.

Because the data set provides information at a country level, we can think of each country as an individual market. This means we can frame our goal as finding the two best countries to advertise in.

One indicator of a good market is the number of potential customers — the more potential customers in a market, the better. If our ads manage to convince 10% of the 5000 potential customers in market A to buy our product, then this is better than convincing 100% of the 30 potential customers in market B.

We'll only consider participants who actually indicated what role(s) they'd be interested in, to make sure we're working with a representative sample:

```python
data_with_interests = survey_data[survey_data["JobRoleInterest"].notnull()].copy()
```

```python
# Generate both absolute numbers and percentage values
abs_freq = data_with_interests["CountryLive"].value_counts()
pct_freq = data_with_interests["CountryLive"].value_counts(normalize = True) * 100

# Visualize using a dataframe for greater visual clarity
pd.DataFrame({"Absolute Frequency": abs_freq, "Percentage": pct_freq})
```

We can see that the United States is, by far, the most interesting market with 45.7% of the total.  India is second, but at only 7.7%, it's not that far ahead of the United Kingdom (4.6%) or Canada (3.8%).

This is useful information, but we need to go more in-depth and understand how much money people in these countries are actually willing to spend on learning.  Advertising within markets where most people are only willing to learn for free is extremely unlikely to be profitable for us.

## Spending Money For Learning

The `MoneyForLearning` column describes, in USD, the amount of money spent by participants from the moment they started coding until the moment they completed the survey. Our company sells subscriptions at a price of $59 per month, and for this reason we're interested in finding out how much money each student spends per month.

It also seems like a good idea to narrow down our analysis to only four countries: the US, India, the United Kingdom, and Canada. Two reasons for this decision are:

- These are the countries that have the highest absolute frequencies in our sample, which means we have a decent amount of data for each.
- Our courses are written in English, and English is an official language in all four of these countries. The more people that know English, the better our chances are of targeting the right people with our ads.

(Note: Some students answered that they had been learning to code for 0 months [it might be that they had just started when they completed the survey]. For the sake of our analysis, we're going to replace these 0s with 1s.)

```python
data_with_interests["MonthsProgramming"].replace(0, 1, inplace = True)
```

```python
# Calculate spend-per-month by dividing total money spent by months programming
data_with_interests["SpentPerMonth"] = data_with_interests["MoneyForLearning"] / data_with_interests["MonthsProgramming"]
```

```python
data_with_interests["SpentPerMonth"].isnull().sum() # Check how many resulting values are null
```

We're only going to keep/use the rows that don't have null values for the `SpentPerMonth` column.

```python
data_with_interests = data_with_interests[data_with_interests["SpentPerMonth"].notnull()]
```

We're also going to eliminate rows that nave null values for the `CountryLive` column, since these aren't useful to us, either.  We'll double-check that we still have enough data for the four countries that interest us.

```python
data_with_interests = data_with_interests[data_with_interests["CountryLive"].notnull()]
data_with_interests["CountryLive"].value_counts().head()
```

This should still be enough, so let's determine the average amount spent per month by a participant in each country.  We'll determine the average using the mean.

```python
countries_mean = data_with_interests.groupby("CountryLive").mean()

# Limit result to our 4 primary countries of interest
countries_mean["SpentPerMonth"][["United States of America", "India", "United Kingdom", "Canada"]]
```

The results for the United Kingdom and Canada are surprisingly low relative to the values we see for India. If we considered a few socio-economical metrics (like [GDP per capita](https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)_per_capita#Lists_of_countries_and_dependencies)), we'd intuitively expect people in the UK and Canada to spend more on learning than people in India.

It might be that we don't have have enough representative data for the United Kingdom, Canada, and India, or we have some outliers (maybe coming from wrong survey answers) making the mean too big for India, or too low for the UK and Canada. Or it might be that the results are correct.

## Dealing With Extreme Outliers

We'll use box plots to visualize, for each of our countries of interest, the distribution of the `SpentPerMonth` variable.

```python
# Isolate our countries of interest
four_countries = data_with_interests[data_with_interests["CountryLive"].str.contains("United States of America|India|United Kingdom|Canada")]
```

```python
import seaborn as sns
sns.boxplot(y = "SpentPerMonth", x = "CountryLive", data = four_countries)
plt.title("Money Spent Per Month Per Country (Distributions)", y = 1.1)
plt.ylabel("Money Spent Per Month (USD)")
plt.xlabel("Country")
plt.xticks(range(4), ["US", "UK", "India", "Canada"])
plt.show()
```

It is difficult to identify whether there are extreme outliers for the UK, India, or Canda; but there are two very clear ones for the United States (two people spending at least \\$50,000 per month).  This certainly isn't impossible, but is unlikely, so we will remove every value that goes over $20,000 per month.

```python
data_with_interests = data_with_interests[data_with_interests["SpentPerMonth"] < 20000]
```

Now we'll recompute mean values and take a look at our distribution box plots again:

```python
countries_mean = data_with_interests.groupby("CountryLive").mean()
countries_mean["SpentPerMonth"][["United States of America", "India", "United Kingdom", "Canada"]]
```

```python
# Isolate our countries of interest again
four_countries = data_with_interests[data_with_interests["CountryLive"].str.contains("United States of America|India|United Kingdom|Canada")]

# Re-visualize with box plots
sns.boxplot(y = "SpentPerMonth", x = "CountryLive", data = four_countries)
plt.title("Money Spent Per Month Per Country (Distributions)", y = 1.1)
plt.ylabel("Money Spent Per Month (USD)")
plt.xlabel("Country")
plt.xticks(range(4), ["US", "UK", "India", "Canada"])
plt.show()
```

Also of note are a couple of extreme outliers for India (specifically, values of $2500+).  It's unclear whether this is good data or not, as the possibility exists these are individuals who attended bootcamps.  We'll take a closer look at these data points to see if we notice anything relevant.

```python
outliers_india = four_countries[(four_countries["CountryLive"] == "India") & (four_countries["SpentPerMonth"] >= 2500)]
outliers_india
```

Based on the above data, it doesn't appear as though any of these outlier individuals attended bootcamps.  Overall, it's unclear as to why these individuals would have spent this much on learning.  It's possible that these are numbers related to university tuition, which falls outside the scope of our analysis (we're looking to establish amounts of money spent on learning costs _other_ than university tuition).  That said, we'll remove these rows, as that seems the safer choice.

```python
four_countries = four_countries.drop(outliers_india.index)
```

When we look back at our box plots, we also see several extreme outliers for the United States (values over $6000).  Let's take a closer look at these, as well:

```python
outliers_us = four_countries[(four_countries["CountryLive"] == "United States of America") & (four_countries["SpentPerMonth"] >= 6000)]
outliers_us
```

We can see a couple of interesting things here:

- Out of these outliers, 6 people attended bootcamps, which can explain the large amounts of money dedicated to learning.  For the other 5, it's difficult to tell why their money spent amounts are this high, so we'll remove them.
- There are 8 individuals who indicated 3 or fewer months programming.  It's likely that they paid a significant amount of money up front to a bootcamp, affecting their total amount spent and thus causing our `SpentPerMonth` calculation to give them a \[likely unrealistic\] per-month value.  We'll remove these as well.

```python
# Drop non-Bootcamp attendees from above
us_high_no_bootcamp = four_countries[
    (four_countries["CountryLive"] == "United States of America") &
    (four_countries["SpentPerMonth"] >= 6000) &
    (four_countries["AttendedBootcamp"] == 0)
]
four_countries = four_countries.drop(us_high_no_bootcamp.index)

# Drop individuals from above with 3 months or fewer programming
us_three_months_or_fewer = four_countries[
    (four_countries["CountryLive"] == "United States of America") &
    (four_countries["SpentPerMonth"] >= 6000) &
    (four_countries["MonthsProgramming"] <= 3)
]
four_countries = four_countries.drop(us_three_months_or_fewer.index)
```

Finally, taking one more look at our box plots, we see one extreme outlier right around the $5000 per month mark.  Let's take a closer look:

```python
canada_outliers = four_countries[(four_countries["CountryLive"] == "Canada") & (four_countries["SpentPerMonth"] > 4500)]
canada_outliers
```

This outlier looks similar to those from the United States who appear to have paid a large lump sum up front, but have only been programming for a couple of months, causing the per-month value to be unrealistically high.  We'll take the same approach as we did for these outliers in the US and remove the row.

```python
four_countries = four_countries.drop(canada_outliers.index)
```

Now let's re-calculate mean values for `SpentPerMonth`.

```python
four_countries.groupby("CountryLive").mean()["SpentPerMonth"]
```

```python
# Visualize one more time with box plots
sns.boxplot(y = "SpentPerMonth", x = "CountryLive", data = four_countries)
plt.title("Money Spent Per Month Per Country (Distributions)", y = 1.1)
plt.ylabel("Money Spent Per Month (USD)")
plt.xlabel("Country")
plt.xticks(range(4), ["US", "UK", "India", "Canada"])
plt.show()
```

## Choosing the Two Best Markets

Considering the results we've found so far, one country we should definitely advertise in is the U.S.  There are a lot of new coders living there and they are willing to pay a good amount of money each month.

As for the second country, there are couple of potential strategies:

- Focus on India, as it has the second largest population of new coders (based on the data) and, when considering that multipled by the mean per-month spend by individuals in India, creates a greater total than applying the same strategy to Canada (which has a greater mean per-month spend value).  India as a country also has a very large population, providing for a a substantial potential advertising reach.
- Focus on Canada based on the second highest per-month mean money spend, with the idea that, with engagement through advertising, we can take greater advantage of that spend.

It may also be potentially beneficial to advertise in more than two countries.  It may behoove us to experiment with advertising in both India and Canada (on top of the United States) to get an idea what our direct advertising results would be.

At this point, it's probably best to send our analyss to our marketing team and give them an opportunity to use their domain knowledge to decide.  They might want to do some further information gathering when it comes to India and Canada, in particular (i.e. through further surveys).

In this project, we analyzed survey data from new coders to find the best two markets to advertise in. The only solid conclusion we reached is that the US would be a good market to advertise in.

For the second best market, it wasn't clear-cut what to choose between India and Canada. We decided to send the results to the marketing team so they can use their domain knowledge to take the best decision.
